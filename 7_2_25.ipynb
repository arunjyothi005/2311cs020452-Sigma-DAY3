```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

cars[['HP','VOL','SP','WT','MPG']]=sc.fit_transform(cars[['HP','VOL','SP','WT','MPG']])
```

```python
np.round(cars.describe(),3)
```

```python
import statsmodels.formula.api as smf

ml_vol=smf.ols('MPG~VOL',data=cars).fit()
np.round(ml_vol.rsquared,2)
```

```python
ml_vol.params
```

```python
import statsmodels.formula.api as smf

ml_vol=smf.ols('HP~VOL',data=cars).fit()
np.round(ml_vol.rsquared,2)
```

```python
import statsmodels.formula.api as smf

ml_vol=smf.ols('SP~VOL',data=cars).fit()
np.round(ml_vol.rsquared,2)
```

```python
import statsmodels.formula.api as smf

model=smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
np.round(model.rsquared,2)
```
